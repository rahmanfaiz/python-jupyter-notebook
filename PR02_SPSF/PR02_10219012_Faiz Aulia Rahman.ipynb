# Soal
Sebuah partikel bermuatan q bergerak dengan kecepatan,
$\vec{\nu}(t) = \nu_{x}(t)\hat{i}+\nu_{y}(t)\hat{j}+$ 
berada dalam suatu ruang bermedan magnetik konstan $\vec{B}= - \hat{k}B_{z}$
akan ditentukan:
<br>
<br>
a. persamaan dari hukum Newtonnya
<br>
b. persamaan diferensial terkopel antara kecepatan kedua arah
<br>
c. penyelesaian kedua persamaan diferensial sehingga diperoleh $\nu_{x}(t),\nu_{y}(t),x(t),$ dan $y(t)$
<br>
d. solusi numeriknya
<br>
e. perbandingan antara hasil dari teori dan numerik

# Jawaban
a. persamaan dari hukum Newtonnya
<br>
<br>
$$\Sigma F = ma$$
<br>
$$F = q\nu \times B = q(\nu_{x}(t)\hat{i}+\nu_{y}(t)\hat{j})\times(-\hat{k}B_{z})$$
<br>
$$ma = q(-\nu_{y}\hat{i}+\nu_{x}\hat{j})B_{z}$$
<br>
$$\frac{d\vec{\nu}}{dt} = \frac{q(-\nu_{y}\hat{i}+\nu_{x}\hat{j})B_{z}}{m}$$

b. persamaan diferensial terkopel antara kecepatan kedua arah
<br>
dengan $\omega=\frac{qB_{z}}{m}$ maka,
<br>
$$\frac{d\nu_{x}}{dt} = -\omega \nu_{y}(t)$$
<br>
$$\frac{d\nu_{y}}{dt}=\omega \nu_{x}(t)$$

c. penyelesaian kedua persamaan diferensial sehingga diperoleh $\nu_{x}(t),\nu_{y}(t),x(t),$ dan $y(t)$
<br>
tinjau turunan kedua dari $\frac{d\nu_{x}}{dt}$ dan $\frac{d\nu_{y}}{dt}$
$$\frac{d^{2}\nu_{x}}{dt^{2}} = -\omega\frac{d\nu_{y}}{dt}$$
<br>
masukkan persamaan dari bagian sebelumnya
<br>
$$\frac{d^{2}\nu_{x}}{dt^{2}} = -\omega^{2}\nu_{x}$$
<br>
sehingga solusinya,
<br>
$$\nu_{x}(t)= A\sin{\omega t}$$
<br>
kemudian, dilakukan hal yang sama dengan persamaan diatas
<br>
<br>
$$\frac{d^{2}\nu_{y}}{dt^{2}}=\omega\frac{d\nu_{x}}{dt}$$
<br>
$$\frac{d^{2}\nu_{y}}{dt^{2}} = -\omega^{2}\nu_{y}$$
<br>
sehingga solusinya,
<br>
$$\nu_{y}(t)= - A\cos{\omega t}$$

d. solusi numeriknya
<br>
dimasukkan persamaan yang telah didapatkan,
<br>
<br>
$$\frac{d\nu_{x}}{dt} = -\omega \nu_{y}(t)$$
<br>
$$\frac{d\nu_{y}}{dt}=\omega \nu_{x}(t)$$
<br>
dengan $\omega=\frac{qB_{z}}{m}$ dan nilai $\omega = 1$
akan didapatkan

```python
import numpy as np
import matplotlib.pyplot as plt

## fungsi-fungsi ##

def run_kut4(F,t,x,y,h):
    K0=h*F(t,x,y)
    K1=h*F(t + h/2.0, x + K0/2.0, y)
    K2=h*F(t + h/2.0, x + K1/2.0, y)
    K3=h*F(t + h, x + K2, y)
    return (K0 + 2.0*K1 + 2.0*K2 +K3)/6.0

def vxdot(t, vx, vy, c=1):
    return -c*vy

def vydot(t, vy, vx, c=1):
    return c*vx


#####################

vx0 = 0 
vy0 = 1 
t0 = 0
tStop = 30
h = 0.005

vxs=[vx0]
vys=[vy0]
ts=[t0]

while ts[-1] < tStop:
    h = min(h, tStop - ts[-1])
    vx = vxs[-1] + run_kut4(vxdot,ts[-1],vxs[-1],vys[-1],h)
    vy = vys[-1] + run_kut4(vydot,ts[-1],vys[-1],vxs[-1],h)
    t = ts[-1] + h
    ts.append(t)
    vxs.append(vx)
    vys.append(vy)
  
plt.figure()
plt.plot(ts,vys,ts,vxs)
plt.xlabel('Waktu')
plt.ylabel('Kecepatan')
plt.legend(['$v_y$', '$v_x$'])

plt.show()
```

e. perbandingan antara hasil dari teori dan numerik
